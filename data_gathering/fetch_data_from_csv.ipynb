```python
import requests
from io import StringIO
import pandas as pd
```

```python
url = 'https://www.w3schools.com/python/pandas/data.csv.txt'
#headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url)

# Manually convert to dataframe
records = []
for row in req.text.split('\n'):
    records.append(row.split(','))

 # mking 1st row as header row
 # & removing 1st column of nos & last record
df = pd.DataFrame(records[1:-1],columns=records[0])
df
```

```python
# Convert to dataframe using io - StringIO
data = StringIO(req.text)

df = pd.read_csv(data)
df
```
