# Gather Data from UCI Machine Learning Repository

```python
import pandas as pd
import requests
from io import StringIO
```

```python
# converting from .data to pandas dataframe

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
response = requests.get(url)
data = StringIO(response.text)
pd.read_csv(data,names=['sepal length','sepal width', 'petal length','petal width','class'])
```
